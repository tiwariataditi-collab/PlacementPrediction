```python
import pandas as pd
datafr = pd.read_csv('/placement.csv').squeeze()
datafr = datafr.iloc[:,1:]
datafr.head()
```

```python
# steps

# 0 : preprocess + EDA + feature seclection
# 1 : extract input and outpur cols
# 2 : scales the value
# 3 : train the test split
# 4 : train the model
# 5 : evaluate the model
# 6 : deploy the model
```

```python
import matplotlib.pyplot as plt
plt.scatter(datafr['cgpa'], datafr['iq'], c=datafr['placement'])
```

```python
x = datafr.iloc[:, 0:2]
x
```

```python
x.shape
```

```python
y = datafr.iloc[:,-1]
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x_train, y_train.values, clf=clf , legend = 2)
plt.show()
```
